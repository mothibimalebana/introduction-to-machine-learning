## Titanic Survival Prediction Dataset:
#### PassengerId: Passenger ID


#### Survived: Weather Survived or not: 0 = No, 1 = Yes


#### Pclass: Ticket class: 1 = 1st, 2 = 2nd, 3 = 3rd


#### Name: Name of the Passenger


#### Sex: Gender


#### Age: Age in Years


#### SibSp: No. of siblings / spouses aboard the Titanic


#### Parch: No. of parents / children aboard the Titanic


#### Ticket: Ticket number


#### Fare: Passenger fare


## step 1: load data using pandas

```python
import dask.dataframe as dd
import pandas as pd
from dask.distributed import Client, LocalCluster
from dask_ml.model_selection import train_test_split
from dask_ml.ensemble import RandomForestClassifier
from dask_ml.metrics import accuracy_score
import time


# Load the data using pandas
df = pd.read_csv('titanic3.csv')
```

## step 2: cleaning the data

```python
# Cleaning the data
df['age'] = df['age'].fillna(df['age'].mean().astype(int)) #handle missing values for age
df['fare'] = df['fare'].fillna(df['fare'].mean().astype(int)) #handle missing values for fare
df = pd.get_dummies(df, columns=['sex'], drop_first=True, dtype=int) #handle categorical variables

# Keep significant features
keep_cols = ['pclass', 'sex_male', 'age', 'sibsp', 'parch', 'fare', 'survived']
df_clean = df[keep_cols]
df = dd.from_pandas(df_clean, npartitions=4)
print(f"   Dask type: {type(df)}")
print(f"   Partitions: {df.npartitions}")
print(f"   Still lazy? {not hasattr(df, 'compute')}")
```

### step 3: dask local cluster

```python
cluster = LocalCluster(n_workers=2, threads_per_worker=2)
client = Client(cluster)
print(f"   Dashboard: {client.dashboard_link}")
print(f"   Workers: {len(cluster.workers)}")
```
