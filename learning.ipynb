## Exercise 1.2: Functions
## Task: Write a function that takes a list and returns only even numbers

```python
def get_even(list):
    even_list = []
    for number in list:
        if number % 2 == 0:
            even_list.append(number)
        else:
            continue
    return even_list

print(get_even(numbers))
```

## Exercise 1.3: Dictionary Operations
### Task: Count how many times each word appears in a sentence

```python
def word_count(sentence):
    sentence_array = sentence.rsplit(" ")
    words = {}
    for word in sentence_array:
        if words.get(word) == None:
            words[word] = 1
        else:
            words[word] += 1
    return words


        
words = {}
print(word_count("Hello how are you doing Are you good"))
```
