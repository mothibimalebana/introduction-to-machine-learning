## Create a list of genders

```python
genders = ['male', 'female', 'male', 'male', 'female', 'female']
adjusted_genders = []
for person in genders:
    if person == 'male':
        adjusted_genders.append(1)
    else:
        adjusted_genders.append(0)
print(adjusted_genders)
```

# DataFrame with missing ages


```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'age': [25, np.nan, 30, np.nan, 35],
    'city': ['NYC', 'LA', 'NYC', 'Chicago', 'LA']
})

median_age = df['age'].median()

for age in df['age']:
    if df['age'].isnull().any():
        df['age'] = df['age'].fillna(median_age)
    else:
        continue
```
